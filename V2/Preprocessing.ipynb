```python
from PyPDF2 import PdfReader
import os

pdf_file_paths = [r"D:\Graduation_project\Diagnostic and statistical manual of mental disorders _ DSM-5 ( PDFDrive.com ).pdf",r"D:\Graduation_project\19-mh-8084-autismspectrumdisorder.pdf",r"D:\Graduation_project\اضطرابات نفسيه وعقليه .pdf",r"D:\Graduation_project\يوسف علاء .pdf",r"D:\Graduation_project\ذوي_الاحتياجات_الخاصة[1].pdf"]
pdf_pages = []
for pdf_file_path in pdf_file_paths:
    with open(pdf_file_path, 'rb') as pdf_file:
        reader = PdfReader(pdf_file)
        for page_number in range(len(reader.pages)):
            for content in reader.pages[page_number].extract_text().split('.\n'):
                if len(content.split()) >= 10:
                    pdf_page_info = {
                        'information': content,
                        'source_file_name': os.path.basename(pdf_file_path),
                        'page_number': page_number + 1
                    }
                    pdf_pages.append(pdf_page_info)

pdf_pages[-10]
```
